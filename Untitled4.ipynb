```python
# Task 1: Prompt the user to input two integers
num1 = int(input("Enter the first integer: "))
num2 = int(input("Enter the second integer: "))
```

```python
# Task 2: Calculate the sum of the two integers
sum_of_numbers = num1 + num2
print("The sum of", num1, "and", num2, "is:", sum_of_numbers)
```

```python
# Task 3: Check if the computed sum lies between 10 and 20 (inclusive)
if 10 <= sum_of_numbers <= 20:
    print("Output: 15")
else:
    print("Output: 0")
```

```python
# Task 4: Provide informative feedback
if 10 <= sum_of_numbers <= 20:
    print("The sum lies between 10 and 20, so the result is set to 15.")
else:
    print("The sum is outside the range 10-20, so the result is set to 0.")
```
