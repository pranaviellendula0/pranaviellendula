```python
# Prompt the user to input a number and convert it to an integer
number = int(input("Enter the number for which you want the multiplication table: "))

# Print a header for the table
print(f"\nMultiplication Table for {number} (1 to 10):\n")

# Use a for loop to iterate from 1 to 10 (range is exclusive of the end, so we use 11)
for i in range(1, 11):
    # Calculate the product
    product = number * i

    # Print the result in the format: number x i = product
    print(f"{number} x {i} = {product}")
```
