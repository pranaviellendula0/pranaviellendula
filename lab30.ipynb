```python
# Prompt the user for a starting and ending number
start = int(input("Enter the starting number for the range: "))
end = int(input("Enter the ending number for the range: "))

# Initialize the first two Fibonacci numbers
# F0 = 0, F1 = 1
a = 0
b = 1

# List to store Fibonacci numbers within the range
fibonacci_in_range = []

# Generate Fibonacci numbers using a while loop
while a <= end:
    # Check if the current Fibonacci number is within the range (inclusive of start)
    if a >= start:
        fibonacci_in_range.append(a)

    # Calculate the next Fibonacci number
    next_fib = a + b

    # Update a and b for the next iteration
    a = b
    b = next_fib

# Check if any numbers were found and print the result
if fibonacci_in_range:

    print(f"\nFibonacci numbers between {start} and {end} are: {fibonacci_in_range}")
else:
    print(f"\nNo Fibonacci numbers found in the range {start} to {end}.")
```
