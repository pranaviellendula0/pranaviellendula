```python
temp = input("Input the temperature you'd like to convert (e.g., 60F, 45C): ")

# Extract the numeric part of the input (all but the last character)
degree = int(temp[:-1])

# Extract the unit (the last character) and convert it to uppercase
unit = temp[-1].upper()

if unit == "C":
    # Formula: F = (C * 9/5) + 32
    fahrenheit = (degree * 9/5) + 32
    print(f"The temperature in Fahrenheit is {fahrenheit} F")

elif unit == "F":
    # Formula: C = (F - 32) * 5/9
    celsius = (degree - 32) * 5/9
    print(f"The temperature in Celsius is {celsius} C")

else:
    print("Invalid temperature format. Please use C or F (e.g., 60F, 45C).")
```
