```python
# Task 1: Implement the Python program to convert a month name to a number of days
month = input("Input the name of Month: ").lower()

if month in ('january', 'march', 'may', 'july', 'august', 'october', 'december'):
    print("Number of days: 31")

elif month in ('april', 'june', 'september', 'november'):
    print("Number of days: 30")

elif month == 'february':
    # Note: The problem statement says leap year handling is not required,
    # but the code outputs both possibilities.
    print("Number of days: 28/29")

else:
    print("Invalid month name")
```
