```python
my_tuple=(50,60,70,80,90)
print("tuple:",my_tuple)
```

```python
print("first element:",my_tuple[0])
print("last element:" ,my_tuple[3])
```

```python
print("slice [1:4]", my_tuple[1:4])
print("slice from beginning to index 3:" , my_tuple[:3])
print("slice from index 2 to end:" ,my_tuple[2:])
```

```python
extra_tuple=(100,110,120)
concatenated=my_tuple + extra_tuple
print("concatenateed tuple:", concatenated)
```

```python
for item in my_tuple:
  print("item:" ,item)
```

```python
print("is 70 in tuple?" ,70 in my_tuple)
print("is 100 not in tuple?",100 not in my_tuple)
```
