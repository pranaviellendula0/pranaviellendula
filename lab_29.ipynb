```python
# Prompt the user for a starting and ending number
start = int(input("Enter the starting number: "))
end = int(input("Enter the ending number: "))

# Check if the starting number is less than the ending number
if start < end:
    print(f"\nNumbers between {start} and {end} divisible by 3 and multiples of 5 are:")

    # Iterate through the range (note: range is exclusive of the end number)
    for num in range(start, end):
        # Check if the number is divisible by 3 (num % 3 == 0) AND a multiple of 5 (num % 5 == 0)
        if num % 3 == 0 and num % 5 == 0:
            print(num)
else:
    print("Invalid input. The starting number must be less than the ending number.")

```
