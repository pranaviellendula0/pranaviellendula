```python
# Initialize a counter
counter = 1

# Loop while counter is less than or equal to 5
while counter <= 5:
    print(counter)
    counter += 1
```

```python
# Correct number to guess
correct_number = 7

# Initialize the user's guess
guess = 0

# Keep asking the user to guess until they get it right
while guess != correct_number:
    guess = int(input("Guess the number between 1 and 10: "))

print("Congratulations! You guessed the correct number.")
```

```python
# Start at 10
number = 10

# Decrement the number until it reaches 5, then break the loop
while number > 0:
    print(number)
    number -= 1
    if number == 5:
        break

print("Loop terminated early because the number reached 5.")
```

```python
# Initialize a number
number = 0

# Loop while number is less than 10
while number < 10:
    number += 1
    if number % 2 == 0:
        continue # Skip the rest of the loop for even numbers
    print(number)
```

```python
# Start an infinite loop
while True:
    name = input("Enter your name (type 'exit' to stop): ")
    if name.lower() == 'exit':
        break
    print(f"Hello, {name}!")

print("Loop exited.")
```
