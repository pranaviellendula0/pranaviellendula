```python
# Create a list of fruits
fruits = ["apple", "banana", "cherry", "date"]
print(f"Original List: {fruits}")
```

```python
# Access elements
print(f"First fruit: {fruits[0]}")
print(f"Last fruit: {fruits[-1]}")
```

```python
# Slicing the list
sub_list = fruits[1:3]
print(f"Sliced List (index 1 to 2): {sub_list}")
```

```python
# Modify an element
fruits[1] = "blackberry"
print(f"Modified List: {fruits}")
```

```python
# Append an element to the end
fruits.append("elderberry")
print(f"List after append: {fruits}")

# Insert an element at a specific index
fruits.insert(2, "fig")
print(f"List after insert: {fruits}")
```

```python
# Remove a specific element
fruits.remove("date")
print(f"List after remove('date'): {fruits}")
```

```python
# Pop the last element
popped_fruit = fruits.pop()
print(f"Popped fruit: {popped_fruit}")
print(f"List after pop: {fruits}")
```

```python
# Length of the list
list_length = len(fruits)
print(f"Length of the list: {list_length}")
```

```python
# Check for membership
check_fruit = "apple"
if check_fruit in fruits:
    print(f"'{check_fruit}' is in the list.")
else:
    print(f"'{check_fruit}' is not in the list.")
```

```python
# Iterate over the list
print("Iterating through fruits:")
for fruit in fruits:
    print(fruit)
```

```python
# Sort the list alphabetically (in place)
fruits.sort()
print(f"Sorted List: {fruits}")
```

```python
# Reverse the list (in place)
fruits.reverse()
print(f"Reversed List: {fruits}")
```

```python
# Example of List Comprehension: Create a list of lengths for each fruit
fruit_lengths = [len(fruit) for fruit in fruits]
print(f"List Comprehension (lengths): {fruit_lengths}")
```

```python
# Copying a list (using slicing)
fruits_copy = fruits[:]
fruits_copy.append("guava")
print(f"Original list remains: {fruits}")
print(f"Copied list: {fruits_copy}")
```

```python
# Clearing a list
fruits.clear()
print(f"List after clear: {fruits}")
```
