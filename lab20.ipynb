```python



#identity comparison
x=[4,5,6]
y=[11,12,13]
z=x
print(f"x is y:{x is y}")
print(f"x is z:{x is z}")
print(f"x is not y:{x is not y}")
print(f"x is not z:{x is not z}")
```

```python
#membership operators
amulya=[1,2,3,4,5]
pranavi=4
sneha_yatala=5
print(f"{pranavi} in amulya:{pranavi in amulya}")
print(f"{sneha_yatala} in amulya:{sneha_yatala in amulya}")
print(f"{pranavi} not in amulya:{pranavi not in amulya}")
print(f"{sneha_yatala} not in amulya:{sneha_yatala not in amulya}")
```
